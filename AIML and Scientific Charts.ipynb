```python
import plotly.graph_objects as go
import numpy as np
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Load and split data
X, y = make_moons(noise=0.3, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(
    X, y.astype(str), test_size=0.25, random_state=0)

trace_specs = [
    [X_train, y_train, '0', 'Train', 'square'],
    [X_train, y_train, '1', 'Train', 'circle'],
    [X_test, y_test, '0', 'Test', 'square-dot'],
    [X_test, y_test, '1', 'Test', 'circle-dot']
]

fig = go.Figure(data=[
    go.Scatter(
        x=X[y==label, 0], y=X[y==label, 1],
        name=f'{split} Split, Label {label}',
        mode='markers', marker_symbol=marker
    )
    for X, y, label, split, marker in trace_specs
])
fig.update_traces(
    marker_size=12, marker_line_width=1.5,
    marker_color="lightyellow"
)
fig.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

mesh_size = .02
margin = 1

# We will use the iris data, which is included in px
df_train, df_test = train_test_split(df, test_size=0.25, random_state=0)
X_train = df_train[['How_much_time_are_you_sleeping_everyday', 'What_is_your_screen_time']]
y_train = df_train.concentrate_on_your_work_when_you_are_stressed

# Create a mesh grid on which we will run our model
l_min, l_max = df.How_much_time_are_you_sleeping_everyday.min() - margin, df.How_much_time_are_you_sleeping_everyday.max() + margin
w_min, w_max = df.What_is_your_screen_time.min() - margin, df.What_is_your_screen_time.max() + margin
lrange = np.arange(l_min, l_max, mesh_size)
wrange = np.arange(w_min, w_max, mesh_size)
ll, ww = np.meshgrid(lrange, wrange)

# Create classifier, run predictions on grid
clf = KNeighborsClassifier(15, weights='distance')
clf.fit(X_train, y_train)
Z = clf.predict(np.c_[ll.ravel(), ww.ravel()])
Z = Z.reshape(ll.shape)
proba = clf.predict_proba(np.c_[ll.ravel(), ww.ravel()])
proba = proba.reshape(ll.shape + (2,))

# Compute the confidence, which is the difference
diff = proba.max(axis=-1) - (proba.sum(axis=-1) - proba.max(axis=-1))

fig = px.scatter(
    df_test, x='How_much_time_are_you_sleeping_everyday', y='What_is_your_screen_time',
    symbol='concentrate_on_your_work_when_you_are_stressed',
    symbol_map={
        'Yes': 'square-dot',
        'No': 'circle-dot'},
)
fig.update_traces(
    marker_size=12, marker_line_width=1.5,
    marker_color="lightyellow"
)
fig.add_trace(
    go.Heatmap(
        x=lrange,
        y=wrange,
        z=diff,
        opacity=0.25,
        customdata=proba,
        colorscale='RdBu',
        hovertemplate=(
            'How_much_time_are_you_sleeping_everyday: %{x} <br>'
            'What_is_your_screen_time: %{y} <br>'
            'p(Yes): %{customdata[0]:.3f}<br>'
            'p(No): %{customdata[1]:.3f}<br>'
        )
    )
)
fig.update_layout(
    legend_orientation='h',
    title='Prediction Confidence on Test Split'
)
fig.show()
```

```python
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sklearn.svm import SVR

mesh_size = .02
margin = 0

df = pd.read_csv('Survey on Student Mental Health1 (1).csv')
X = df[['How_much_time_are_you_sleeping_everyday','What_is_your_screen_time']]
y = df['Current_year_you_are_studying_in']

# Condition the model on How_much_time_are_you_sleeping_everyday and length, predict the petal width
model = SVR(C=1.)
model.fit(X, y)

# Create a mesh grid on which we will run our model
x_min, x_max = X.How_much_time_are_you_sleeping_everyday.min() - margin, X.How_much_time_are_you_sleeping_everyday.max() + margin
y_min, y_max = X.What_is_your_screen_time.min() - margin, X.What_is_your_screen_time.max() + margin
xrange = np.arange(x_min, x_max, mesh_size)
yrange = np.arange(y_min, y_max, mesh_size)
xx, yy = np.meshgrid(xrange, yrange)

# Run model
pred = model.predict(np.c_[xx.ravel(), yy.ravel()])
pred = pred.reshape(xx.shape)

# Generate the plot
fig = px.scatter_3d(df, x='How_much_time_are_you_sleeping_everyday', y='What_is_your_screen_time', z='Current_year_you_are_studying_in')
fig.update_traces(marker=dict(size=5))
fig.add_traces(go.Surface(x=xrange, y=yrange, z=pred, name='pred_surface'))
fig.show()
```
